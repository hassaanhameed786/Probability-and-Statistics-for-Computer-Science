```python

import matplotlib 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
url = "https://raw.githubusercontent.com/kshedden/statswpy-nhanes/master/merged/nhanes_2015_2016.csv"
da = pd.read_csv(url)
```

```python
wt = da['BMXWT']
```

```python
plt.figure(figsize=(20, 5)) 

x =  range(da['BMXWT'].size)
y =  da['BMXWT']

plt.scatter(x, y, marker='x', s=0.5)

plt.title("Weights of Individuals")
plt.xlabel("ID of Individual")
plt.ylabel("Weight (kg)")

# plot the estimate line 
p1 = (0, 75)
p2 = (6000, 75)
x_c, y_c = zip(p1, p2)
plt.plot(x_c, y_c, color='red', linewidth=0.95)
plt.show()
```

```python
wt.mean()
```

```python
plt.figure(figsize=(20,5))
# X axis
x = range(da['BMXWT'].size)
#y axis 
y =  da['BMXWT']

plt.scatter(x, y, marker='x', s=0.5)

plt.title("Weights of Individuals")
plt.xlabel("ID of Individual")
plt.ylabel("Weight (kg)")

## estimate line 

x_c,y_c =zip([0,75] , [6000, 75])
plt.plot (x_c , y_c ,color ='red',linewidth =0.85)


## actual line is 
#p3 =(0,81)
#p4 =(6000,81)

x_c,y_c= ([0,6000] , [wt.mean(), wt.mean()])
plt.plot (x_c , y_c ,color ='green',linewidth =0.85)
```
