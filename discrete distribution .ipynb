```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
import seaborn as sns
sns.set(color_codes=True)

import datetime
```

```python
from scipy.stats import bernoulli
```

```python
data_bernouli = bernoulli.rvs(size=1000, p =0.6)
```

```python
ax = sns.displot(data_bernouli,bins=100, kde=False )
plt.title("Discrete Distribution Graph")
ax.set(xlabel= 'bernoulli distribution', ylabel='Frequency')
```
